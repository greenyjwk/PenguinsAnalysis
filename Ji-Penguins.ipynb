# Data Visualization <a class="anchor" id="dv"></a>

## Visualizing the Nan from the dataframe

## Data Clearning

```python
penguins = penguins.dropna()
```

```python
penguins.head()
```

## Distribtion of Sex

```python
sns.set(font_scale = 1.2)
sns.histplot(data=penguins, x="sex", binwidth=3, shrink=.8, color = "blue")
```

## Distribtion of Species

```python
sns.set(font_scale = 1.2)
sns.histplot(data=penguins, x="species", binwidth=3, shrink=.8, color = "blue")
```

## Distribtion of Islands

```python
sns.set(font_scale = 1.2)
sns.histplot(data=penguins, x="island", binwidth=3, shrink=.8, color = "blue")
```

## Distribution of Body Mass

```python
sns.set(font_scale = 1.2)
sns.kdeplot(data=penguins, x="body_mass_g")
sns.rugplot(data=penguins, x="body_mass_g")
```

```python
sns.set(font_scale = 1.2)
sns.kdeplot(data=penguins, x="bill_length_mm")
sns.rugplot(data=penguins, x="bill_length_mm")
```

## Distribution of Bill Depth

```python
sns.set(font_scale = 1.2)
sns.kdeplot(data=penguins, x="bill_depth_mm")
sns.rugplot(data=penguins, x="bill_depth_mm")
```

## Relationship between Fliper and Body Mass

```python
sns.set(font_scale = 1.2)
plot = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=8,
               data=penguins,
                  palette='twilight')
plot.set_xlabels('Flipper Length')
plot.set_ylabels('Body Mass')
```

## Relationship between Flipper Length and Bill Length

```python
sns.set(font_scale = 1.2)
sns.scatterplot(
    data=penguins, x="bill_length_mm", y="flipper_length_mm", hue="species", style = "species", sizes=(20, 200), legend="full")
```

```python
sns.set(font_scale = 1.2)
plot = sns.scatterplot(
    data=penguins, x="bill_depth_mm", y="bill_length_mm", hue="species", style = "species", sizes=(20, 200), legend="full")
```

## Relationship between Flipper Length and Bill Depth

```python
sns.set(font_scale = 1.2)
sns.scatterplot(
    data=penguins, x="bill_depth_mm", y="flipper_length_mm", hue="species", style = "species",
    sizes=(20, 200), legend="full")
```

## Habitat Information

```python
island_spec = sns.countplot(data = penguins[penguins["island"]=="Biscoe"], y = "island", hue = "species", palette = 'husl')
island_spec.set(title = "penguins in Biscoe Island")
```

```python
island_spec = sns.countplot(data = penguins[penguins["island"]=="Dream"], y = "island", hue = "species", palette = 'husl')
island_spec.set(title = "penguins in Dream Island")
```

```python
island_spec = sns.countplot(data = penguins[penguins["island"]=="Torgersen"], y = "island", hue = "species", palette = 'husl')
island_spec.set(title = "penguins in Torgersen Island")
```
